## 이상치 처리

```python
for col in know_2020:
        know_2020[col].replace(' ', 0, inplace = True)
```

```python
know_2020.info(verbose=True)
```

```python
# 주관식문항 

str_li =['bq4_1a','bq4_1b','bq4_1c','bq5_2', 'bq18_10','bq20_1','bq26_1']
```

```python
for i in know_2020.columns : 
    if i not in str_li : 
        know_2020[i] = know_2020[i].astype(int)
    else : 
        know_2020[i] = know_2020[i].astype(str)
```

```python
know_2020.info(verbose=True)
```

```python
## _1에서 1을 선택한 사람중 _2를 답변한 사람이 있는지 확인

for i in range(44) :
    str5 = 'saq' + str(i+1) +'_1'
    str6 = 'saq' + str(i+1) +'_2'
    
    print(know_2020[(know_2020[str5]==1) & (know_2020[str6]!=0)][str6])
    
```
