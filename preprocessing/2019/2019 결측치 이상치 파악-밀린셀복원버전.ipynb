```python
for col in know_2019:
        know_2019[col].replace(' ', 0, inplace = True)
```

```python
# 주관식문항 

str_li =['bq4_1a','bq4_1b','bq4_1c','bq5_2', 'bq18_10','bq20_1','bq22','bq23','bq24', 'bq27_1']
```

```python
know_2019 = know_2019.fillna(0)
```

```python
know_2019.isnull().sum()[100:]
```

```python
# int형으로 바꾸기 위해서 1.null 값처리 2. bq30 번에 37.5 float 형 처리 필요함 

for i in know_2019.columns : 
    if i == 'bq30':
        know_2019[i] = know_2019[i].astype(float).astype(int)
    elif i not in str_li : 
        know_2019[i] = know_2019[i].astype(int)
    else :
        know_2019[i]=know_2019[i].astype(str)
```

```python
know_2019.info(verbose=True)
```

```python
# _1 에서 1을 선택한 사람중 _2 답변을 한 사람이 있는지 확인

for i in range(33) :
    str5 = 'kq' + str(i+1) +'_1'
    str6 = 'kq' + str(i+1) +'_2'
    
    print(know_2019[(know_2019[str5]==1) & (know_2019[str6]!=0)][str6])
    
```
